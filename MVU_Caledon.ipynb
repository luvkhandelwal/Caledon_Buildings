```python
import pandas as pd
```

```python
data= pd.read_csv('Data.csv', encoding_errors= 'replace')
data.head()
```

```python
data.info()
data.dtypes
#data['Subject Details']=str(data['Subject Details'])
#data['Resolution']=str(data['Resolution'])
data['Subject Details']= [str(r) for r in data['Subject Details']]
data['Resolution']= [str(r) for r in data['Resolution']]
print(data['Subject Details'])
```

```python
data.head()
```

```python
data_new=data.groupby('Ticket').agg({
    'Status': 'first',
    'Priority': 'first',
    'Form': 'first',
    'Subject': 'first',
    'Subject Details': ','.join,
    'Resolution':','.join,
    'Ticket Owner': 'first',
    'Submitted': 'first',
    'Due Date': 'first',
    'Technician': 'first',
}).reset_index()
```

```python
data_new.head()
```

```python
data_new.to_csv('Output.csv')
```
